```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('DataFrame_3G_After_Cleaning.csv')
```

```python
corr = df.corr(method='pearson')
ax = plt.subplots(figsize=(20,20), dpi=200)
ax = sns.heatmap(corr, annot=True, linewidths=.5)
```

```python
df1 = df[['CS CSSR', 'CS DROP Rate', 'PS Call Setup SR', 'PS DROP Rate 2']]
corr = df1.corr(method='pearson')
ax = plt.subplots(dpi=200)
ax = sns.heatmap(corr, annot=True)
```

```python
cols = 2
rows = 2
num_cols = df1.select_dtypes(exclude=['object','int64']).columns
fig = plt.figure(figsize=(cols*5,rows*5), dpi=400)
for i, col in enumerate(num_cols) :
    ax = fig.add_subplot(rows, cols, i+1)
    sns.rugplot(x = df1[col], ax=ax, height=.03, color='darkblue')
    sns.histplot(x=df1[col], ax=ax, kde=True).set(title='Distribution des valeurs')
fig.tight_layout()
plt.show()
```

```python
df1.describe()
```

```python
num_cols = df1.columns
for i, col in enumerate(num_cols) :
    skewness = df1[col].skew()
    kurtosis = df1[col].kurt()
    print("Skewness de l'attribut ",col," est : %f"% skewness)
    print("Kurtosis de l'attribut ",col," est : %f"% kurtosis)
```

```python
cols = 2
rows = 2
num_cols = df1.select_dtypes(exclude=['object','int64']).columns
fig = plt.figure(figsize=(cols*5,rows*5), dpi=200)
for i, col in enumerate(num_cols) :
    ax = fig.add_subplot(rows, cols, i+1)
    sns.boxplot(x=df1[col], data=df1, ax=ax)
fig.tight_layout()
plt.show()
```
