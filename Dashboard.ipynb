```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%store -r df
''' CS Drop Rate parameters '''
# IForest
%store -r ifor_nbr_val_sup_threshold
%store -r ifor_nbr_all_anomaly_CSDR
%store -r ifor_nbr_true_anom_CSDR

# Local Outlir Factor
%store -r lof_nbr_all_anomaly_CSDR
%store -r lof_nbr_true_anom_CSDR

# OneClass SVM
%store -r ocsvm_nbr_all_anomaly_CSDR
%store -r ocsvm_nbr_true_anom_CSDR

''' CS Success Rate parameters '''
# IForest
%store -r nbr_val_inf_threshold_CSSR
%store -r ifor_nbr_all_anomaly_CSSR
%store -r ifor_nbr_true_anom_CSSR

# Local Outlier Factor
%store -r lof_nbr_all_anomaly_CSSR
%store -r lof_nbr_true_anom_CSSR

# OneClass SVM
%store -r ocsvm_nbr_all_anomaly_CSSR
%store -r ocsvm_nbr_true_anom_CSSR

''' PS Drop Rate parameters '''
# IForest
%store -r nbr_val_inf_threshold_PSDR
%store -r ifor_nbr_all_anomaly_PSDR
%store -r ifor_nbr_true_anom_PSDR

# Local Outlier Factor
%store -r lof_nbr_all_anomaly_PSDR
%store -r lof_nbr_true_anom_PSDR

# OneClass SVM
%store -r ocsvm_nbr_all_anomaly_PSDR
%store -r ocsvm_nbr_true_anom_PSDR

''' PS Success Rate parameters '''
# IForest
%store -r nbr_val_inf_threshold_PSSR
%store -r ifor_nbr_all_anomaly_PSSR
%store -r ifor_nbr_true_anom_PSSR
# Local Outlier Factor
%store -r lof_nbr_all_anomaly_PSSR
%store -r lof_nbr_true_anom_PSSR
# OneClass SVM
%store -r ocsvm_nbr_all_anomaly_PSSR
%store -r ocsvm_nbr_true_anom_PSSR
```

```python
data = np.array([[ifor_nbr_all_anomaly_CSDR,lof_nbr_all_anomaly_CSDR,ocsvm_nbr_all_anomaly_CSDR],
                [ifor_nbr_all_anomaly_CSSR,lof_nbr_all_anomaly_CSSR,ocsvm_nbr_all_anomaly_CSSR],
                [ifor_nbr_all_anomaly_PSDR,lof_nbr_all_anomaly_PSDR,ocsvm_nbr_all_anomaly_PSDR],
                [ifor_nbr_all_anomaly_PSSR,lof_nbr_all_anomaly_PSSR,ocsvm_nbr_all_anomaly_PSSR]])
x = ['CS Drop Rate', 'CS CSSR', 'PS Drop Rate', 'PS CSSR']
x_axes = np.arange(len(x))
length = len(data)
a = np.arange(length)
width = 0.2
fig,ax = plt.subplots(figsize=(15,10), dpi=300)
plot = plt.bar(a, data[:,0], width, label='Isolation Forest')
for p in plot :
    plt.annotate(p.get_height(), xy=(p.get_x()+0.07, p.get_height()+2), fontsize=12)
plot = plt.bar(a + width, data[:,1], width, label='Local Outlier Factor')
for p in plot :
    plt.annotate(p.get_height(), xy=(p.get_x()+0.07, p.get_height()+2), fontsize=12)
plot = plt.bar(a + (2 * width), data[:,2], width, label='OneClass-SVM')
for p in plot :
    plt.annotate(p.get_height(), xy=(p.get_x()+0.07, p.get_height()+2), fontsize=12)
plt.ylim(-1, 50)
plt.axhline(y=ifor_nbr_val_sup_threshold, color='r', linestyle='dashed')
plt.axhline(y=nbr_val_inf_threshold_CSSR, color='b', linestyle=':')
plt.axhline(y=nbr_val_inf_threshold_PSDR, color='black')
plt.axhline(y=nbr_val_inf_threshold_PSSR, color='y', linestyle='dashed')
plt.xticks(x_axes, x)
plt.xlabel("Features")
plt.ylabel('Nombre des anomalies prédis')
plt.title('Comparaison du resultat des algorithms')
plt.legend(["CS CSDR real anomaly","CS CSSR real anomaly","PS PSSR real anomaly","PS PSDR real anomaly","IForest", "Local utlier Factor", "OC-SVM"])
plt.show()
```
